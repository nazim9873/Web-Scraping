```python
import requests
import re
import csv
from bs4 import BeautifulSoup
```

```python
csv_file=open('Internshala_csv','w')
csv_writer=csv.writer(csv_file)
```

```python
csv_writer.writerow(['job_title','company_name','location','start_date'              
                     ,'duration','stipend','last_date','parttime_allowed'])
```

```python
source=requests.get('https://internshala.com/internships/work-from-home-data%20science-jobs').text
```

```python
soup=BeautifulSoup(source,'lxml')
```

```python
print(soup.prettify())
```

```python
div=soup.find('div',id="internship_list_container_1")
print(div.prettify())
```

```python
job_posting=div.find_all('div',class_='container-fluid individual_internship' or 
               'container-fluid individual_internship visibilityTrackerItem')
```

```python
len(job_posting)
```

```python
y=len(job_posting[0].find_all('a')) #number of a tags in individual job posting
y
```

```python
for i in range(len(job_posting)):
        a_tags=job_posting[i].find_all('a')
        job_title=a_tags[0].text
        job_title=re.sub('^\s$\s','',job_title)
        print(job_title)
        
        company_name=a_tags[1].text
        company_name=re.sub('^\s$\s','',company_name)
        print(company_name)
        
        location=a_tags[2].text
        location=re.sub('^\s$\s','',location)
        print(location)
        
        span_tag=job_posting[i].find_all('span',class_="start_immediately_desktop")
        start_date=span_tag[0].text
        start_date=re.sub('^\s$\s','',start_date)
        
        print(start_date)
        div=job_posting[i].find_all('div',class_='item_body')
        duration=div[1].text
        duration=re.sub('^\s$\s','',duration)
        print(duration)
        
        st_div= job_posting[0].find('span',class_='stipend') 
        stipend=st_div.text
        stipend=re.sub('^\s$\s','',stipend)
        print(stipend)
        
        last_date_container=job_posting[i].find_all('div',class_="item_body")[3]  
        last_date=last_date_container.text
        last_date=re.sub('^\s$\s','',last_date)
        print(last_date)
        
        try:
            if job_posting[i].find_all('div',class_="item_body")[4]:
                    part_time_allowed='yes'
        except Exception as e:
            part_time_allowed='No'
        print(part_time_allowed)  
        csv_writer.writerow([job_title,company_name,location,start_date
                    ,duration,stipend,last_date,part_time_allowed])
        
```

```python
csv_file.close()
```
